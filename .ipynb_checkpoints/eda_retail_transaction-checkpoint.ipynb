## 3. Data Cleaning

```python
print("Dataset info:")
raw_data.info()
```

### 3.1. Rename columns, format data types, remove duplicates

```python

#### Rename columns:
raw_data.rename(columns={'Invoice': 'invoice', 'StockCode': 'stock_code','Description': 'description',
                   'Quantity': 'quantity', 'InvoiceDate': 'invoice_date', 'Price': 'price',
                   'Customer ID': 'customer_id', 'Country':'country'}, inplace=True)

#### Format data types
# InvoiceData
raw_data['invoice_date'] = pd.to_datetime(raw_data['invoice_date'], format='%d.%m.%Y %H:%M')
raw_data['month'] = raw_data['invoice_date'].dt.to_period('M').astype(str)

# CustomerID
raw_data['customer_id'] = raw_data['customer_id'].apply(lambda x: str(int(x)) if pd.notnull(x) else np.nan)

#### Remove duplicates
unique_data = raw_data.drop_duplicates()
unique_data = unique_data.reset_index(drop=True)

print(f'Total rows after removing duplicates {len(unique_data)}')
```

### 3.2. Handle outliers

#### 3.2.1 Quantity

```python
boxplot_plot(unique_data, 'quantity', 'lightskyblue', (3,3), True)
```

```python
# 1. Detect quantity outliers
q1, q3 = unique_data['quantity'].quantile([0.25, 0.75])
iqr = q3 - q1
is_outlier = (unique_data['quantity'] < q1 - 1.5 * iqr) | (unique_data['quantity'] > q3 + 1.5 * iqr)
quantity_outliers = unique_data[is_outlier]

# 2. Identify customers with non-positive total quantity
bad_customers = (
    unique_data.groupby('customer_id')['quantity']
    .sum()
    .reset_index()
    .query('quantity <= 0')['customer_id']
    .unique()
)

# 3. Separate buy and return transactions
df_buy = unique_data[unique_data['quantity'] > 0].copy()
df_return = unique_data[unique_data['quantity'] < 0].copy()
df_return['abs_quantity'] = df_return['quantity'].abs()

# Match returns to buys
br_merged = merge_data(df_buy, df_return, ['customer_id', 'stock_code', 'quantity', 'price'], \
                     ['customer_id', 'stock_code', 'abs_quantity', 'price'], ('_buy', '_return'), 'inner')

br_merged['match_order'] = br_merged.groupby(['customer_id', 'stock_code', 'abs_quantity', 'price']).cumcount() + 1
br_merged = br_merged[br_merged['match_order'] == 1]

# Generate map_id for matched records
br_merged['rm_map1'] = map_id(br_merged, 'customer_id', 'invoice_buy', 'stock_code', 'quantity_buy')
br_merged['rm_map2'] = map_id(br_merged, 'customer_id', 'invoice_return', 'stock_code', 'quantity_return')
unique_data['map_id'] = map_id(unique_data, 'customer_id', 'invoice', 'stock_code', 'quantity')

# 4. Filter out bad customers and matched return-buy pairs
remove_map = set(br_merged['rm_map1']).union(br_merged['rm_map2'])
cleaned_data1 = unique_data[
    (~unique_data['customer_id'].isin(bad_customers)) &
    (~unique_data['map_id'].isin(remove_map))
].drop(columns=['map_id'])

print(f'Total rows after removing QUANTITY outliers: {len(cleaned_data1)}')
```

#### 3.4.2. Price

```python
# 1. Detect and isolate price outliers
q1, q3 = cleaned_data1['price'].quantile([0.25, 0.75])
iqr = q3 - q1
price_outliers = cleaned_data1[
    (cleaned_data1['price'] < q1 - 1.5 * iqr) |
    (cleaned_data1['price'] > q3 + 1.5 * iqr)
]

# 2. Extract invoice type
cleaned_data1['invoice_type'] = (
    cleaned_data1['invoice'].str.extract(r'([A-Za-z]+)')[0]
    .map({'C': 'return', 'A': 'bad_debt'})
    .fillna('purchase')
)

# 3. Remove bad debt records
cleaned_data2 = cleaned_data1[cleaned_data1['invoice_type'] != 'bad_debt'].reset_index(drop=True)

print(f'Total rows after removing PRICE outliers: {len(cleaned_data2)}')
```

#### 3.4.3. Stock code

```python
exclude_exact = ["POST", "DOT", "S", "AMAZONFEE", "CRUK"]
exclude_prefixes = ("BANK CHARGES", "TEST")

# Lọc bỏ các dòng không nằm trong danh sách loại bỏ
cleaned_data = cleaned_data2[
    ~cleaned_data2["stock_code"].isin(exclude_exact) &
    ~cleaned_data2["stock_code"].str.startswith(exclude_prefixes)
]
print(f'Total rows after removing OTHER outliers {len(cleaned_data)}')
```

##### 4.2.4.5. Score products

```python
product_group = identify_dataset.groupby('stock_code').agg(
    total_customers=('customer_id', 'nunique'),
     total_invoices=('invoice', 'nunique'),
    total_revenue=('revenue', 'sum')
).reset_index()
product_group =  product_group[product_group['total_revenue']>0]

product_merge1 = merge_data(product_group, mean_diff_data, 'stock_code', 'stock_code', None, 'left')
product_merge2 = merge_data(product_merge1, longevity_data[['stock_code','longevity_days']], 'stock_code', 'stock_code', None, 'left')
product_merge3 = merge_data(product_merge2, recency_data[['stock_code','recency_days']], 'stock_code', 'stock_code', None, 'left')
product_data = product_merge3

# Normalize data
for col in product_data.columns[1:]:
    product_data[f'{col}_norm'] = normalize_column(product_data, col)
```

```python
def generate_weight_combinations(n_features, step=0.1):
    levels = np.arange(0, 1 + step, step)
    all_combinations = product(levels, repeat=n_features)

    valid_combinations = [list(map(round, comb, [2]*n_features))
                          for comb in all_combinations
                          if abs(sum(comb) - 1.0) < 1e-6]

    return valid_combinations
```

```python
# Danh sách đặc trưng và flag xem có đảo dấu không
FEATURES = ['total_customers_norm', 'total_revenue_norm']
INVERT_FLAGS = [True, False]  # True: cần đảo dấu (1 - x)

def evaluate_combinations(df, features, original_revenue_col, weight_combinations, invert_flags, top_ratio=0.2):
    result_list = []
    df = df.copy()
    for weights in weight_combinations:
        # Tính điểm sản phẩm với phép đảo dấu cho từng đặc trưng nếu cần
        adjusted_features = []
        for feature, invert in zip(features, invert_flags):
            if invert:
                adjusted_features.append(1 - df[feature])
            else:
                adjusted_features.append(df[feature])

        # Kết hợp lại thành DataFrame
        adjusted_matrix = pd.concat(adjusted_features, axis=1)

        # Tính score
        df['product_score'] = adjusted_matrix.dot(weights)

        # Chọn top sản phẩm
        top_n = int(len(df) * top_ratio)
        top_products = df.nlargest(top_n, 'product_score')

        # Tính tỷ lệ doanh thu
        revenue_share = top_products[original_revenue_col].sum() / df[original_revenue_col].sum()

        result_list.append(list(weights) + [revenue_share])

    # Trả về DataFrame kết quả
    columns = [f"w{i+1}" for i in range(len(features))] + ["revenue_share"]
    return pd.DataFrame(result_list, columns=columns)
```

```python
weight_combos = generate_weight_combinations(len(FEATURES), step=0.05)
weight_revenue = evaluate_combinations(product_data, FEATURES, 'total_revenue', weight_combos, INVERT_FLAGS)

# Filter weight has revenue >= 0.6
filtered_weight_revenue = weight_revenue[weight_revenue['revenue_share'] >= 0.5].sort_values(by='revenue_share', ascending=False)
```

```python
# Final score
product_data['score'] = (
    0.15 * (1 - product_data['total_customers_norm'])+
    0.85 * product_data['total_revenue_norm'])
```

```python
def grouped_data(df, group_col, value_col, group_type):
    grouped_df = df.groupby(group_col).agg({value_col: group_type}).reset_index()
    return grouped_df

def generate_customer_features(data):
    df = data.copy()
    ####### Feature engineering
    df_feat_eng = df[['customer_id','label']].drop_duplicates()
    
    ####  1. Customer Country (customer_country)
    # Compute groupby aggregations
    country_df = grouped_data(df, ['customer_id','country'], 'invoice', 'nunique')
    top_country_per_customer = (
      country_df.sort_values('invoice', ascending=False)
      .drop_duplicates(subset='customer_id', keep='first')
    )
    # Merge with feature engineering data
    top_country_per_customer = top_country_per_customer.rename(columns={'country': 'max_country'})
    df_feat_eng = merge_data(df_feat_eng, top_country_per_customer[['customer_id','max_country']], 'customer_id', 'customer_id', None, 'left')
    
    
    ####  2. Number of Invoices per Customer (num_invoices)
    # Compute groupby aggregations
    invoices_df = grouped_data(df[df['invoice_type']=='purchase'], 'customer_id', 'invoice', 'nunique')
    invoices_df = invoices_df.rename(columns={'invoice': 'num_invoice'})
    # Merge with feature engineering data
    df_feat_eng = merge_data(df_feat_eng, invoices_df[['customer_id','num_invoice']], 'customer_id', 'customer_id', None, 'left')
    
    
    ####  3.  Average Monthly Invoices per Customer (avg_monthly_invoices)
    # Compute groupby aggregations
    invoice_per_month = grouped_data(df[df['invoice_type']=='purchase'], ['customer_id', 'month'], 'invoice', 'nunique')
    avg_no_invoice_per_months =  invoice_per_month.groupby('customer_id').agg(avg_monthly_invoices = ('invoice', 'mean')).reset_index()
    # Merge with feature engineering data
    df_feat_eng = merge_data(df_feat_eng, avg_no_invoice_per_months[['customer_id','avg_monthly_invoices']], 'customer_id', 'customer_id', None, 'left')
    

    ####  4. Average Quantity per invoice per customer (avg_invoice_quantity)
    # Compute groupby aggregations
    sum_quantity_per_invoice = grouped_data(df[df['invoice_type']=='purchase'], ['customer_id', 'invoice'], 'quantity', 'sum')
    avg_invoice_quantity = grouped_data(sum_quantity_per_invoice, 'customer_id', 'quantity', 'mean')
    avg_invoice_quantity = avg_invoice_quantity.rename(columns={'quantity': 'avg_invoice_quantity'})
    # Merge with feature engineering data
    df_feat_eng = merge_data(df_feat_eng, avg_invoice_quantity[['customer_id','avg_invoice_quantity']], 'customer_id', 'customer_id', None, 'left')
    
    
    ####  5. Average Product Types per Invoice per Customer (avg_invoice_product_types)
    # Compute groupby aggregations
    stock_code_per_invoice = grouped_data(df[df['invoice_type']=='purchase'], ['customer_id', 'invoice'], 'stock_code', 'nunique')
    avg_invoice_product_types =  stock_code_per_invoice.groupby('customer_id').agg(avg_invoice_product_types = ('stock_code', 'mean')).reset_index()
    # Merge with feature engineering data
    df_feat_eng = merge_data(df_feat_eng, avg_invoice_product_types[['customer_id','avg_invoice_product_types']], 'customer_id', 'customer_id', None, 'left')
    
    
    ####  6. Average Revenue per Invoice per Customer (avg_invoice_revenue)
    # Compute groupby aggregations
    revenue_per_invoice = grouped_data(df, ['customer_id', 'invoice'], 'revenue', 'sum')
    avg_invoice_revenue =  revenue_per_invoice.groupby('customer_id').agg(avg_invoice_revenue = ('revenue', 'mean')).reset_index()
    # Merge with feature engineering data
    df_feat_eng = merge_data(df_feat_eng, avg_invoice_revenue[['customer_id','avg_invoice_revenue']], 'customer_id', 'customer_id', None, 'left')
    
    
    ####  7. Average Purchase Interval per Customer (avg_purchase_interval)
    # Compute groupby aggregations
    invoice_dates = df[df['invoice_type']=='purchase'][['customer_id', 'invoice_date']].drop_duplicates(keep='first')
    invoice_dates['invoice_date'] = pd.to_datetime(invoice_dates['invoice_date'])
    invoice_dates = invoice_dates.sort_values(by=['customer_id', 'invoice_date'])
    invoice_dates['purchase_interval'] = invoice_dates.groupby('customer_id')['invoice_date'].diff().dt.days
    avg_purchase_interval = grouped_data(invoice_dates, 'customer_id', 'purchase_interval', 'mean')
    avg_purchase_interval = avg_purchase_interval.rename(columns={'purchase_interval': 'avg_purchase_interval'})
    # Merge with feature engineering data
    df_feat_eng = merge_data(df_feat_eng, avg_purchase_interval[['customer_id','avg_purchase_interval']], 'customer_id', 'customer_id', None, 'left')
    
    
    ####  8. Average Product Price per Customer (avg_price)
    # Compute groupby aggregations
    stock_code_price_per_customer = df[df['invoice_type']=='purchase'][['customer_id', 'invoice_date','price']].sort_values('invoice_date', ascending=False)
    stock_code_price_per_customer = stock_code_price_per_customer.drop_duplicates(subset='customer_id', keep='first').reset_index(drop=True)
    
    avg_price = grouped_data(stock_code_price_per_customer, 'customer_id', 'price', 'mean')
    avg_price = avg_price.rename(columns={'price': 'avg_price'})
    # Merge with feature engineering data
    df_feat_eng = merge_data(df_feat_eng, avg_price[['customer_id','avg_price']], 'customer_id', 'customer_id', None, 'left')
    
    
    ####  9. Number of Product Types per Customer (num_product_types)
    # Compute groupby aggregations
    num_product_types = grouped_data(df[df['invoice_type']=='purchase'], 'customer_id', 'stock_code', 'nunique')
    num_product_types = num_product_types.rename(columns={'stock_code': 'num_product_types'})
    # Merge with feature engineering data
    df_feat_eng = merge_data(df_feat_eng, num_product_types[['customer_id','num_product_types']], 'customer_id', 'customer_id', None, 'left')
    
    
    ####  10. Revenue per Customer (revenue)
    # Compute groupby aggregations
    revenue = grouped_data(df, 'customer_id', 'revenue', 'sum')
    # Merge with feature engineering data
    df_feat_eng = merge_data(df_feat_eng, revenue[['customer_id','revenue']], 'customer_id', 'customer_id', None, 'left')
    
    return df_feat_eng
```

```python
def check_logit_separation(X, y):
    X = sm.add_constant(X)
    try:
        # Fit mô hình Logit
        model = sm.Logit(y, X)
        result = model.fit()
        return result
    except Exception as e:
        # Nếu gặp phải bất kỳ lỗi khác
        return f"Lỗi: {str(e)}"
```

## 7. RFM Analysis

```python
def categorize(row):
    if row['recency_score'] == 5 and 4 <= row['fm_score'] <= 5:
        return 'Champions'
    elif 3 <= row['recency_score'] <= 4 and 4 <= row['fm_score'] <= 5:
        return 'Loyal Customer'
    elif 4 <= row['recency_score'] <= 5 and 2 <= row['fm_score'] <= 3:
        return 'Potential Loyalist'
    elif row['recency_score'] == 5 and row['fm_score'] == 1:
        return 'New Customers'
    elif row['recency_score'] == 4 and row['fm_score'] == 1:
        return 'Promising'
    elif row['recency_score'] == 3 and 1 <= row['fm_score'] <= 2:
        return 'Needs Attention'
    elif row['recency_score'] == 3 and row['fm_score'] == 3:
        return 'About to Sleep'
    elif 1 <= row['recency_score'] <= 2 and row['fm_score'] == 5:
        return 'Can\'t Loose Them'
    elif 1 <= row['recency_score'] <= 2 and 3 <= row['fm_score'] <= 4:
        return 'At Risk'
    elif 1 <= row['recency_score'] <= 2 and 1 <= row['fm_score'] <= 2:
        return 'Hibernating'
    else:
        return 'Unclassified'
```

```python
# History data
history_data = identify_dataset[
    (identify_dataset['month'].isin(analysis_train_month))
]
history_customers = history_data['customer_id'].unique()

# Test data
labeled_data = identify_dataset[
    (identify_dataset['month'].isin(analysis_test_month)) &
    (identify_dataset['customer_id'].isin(history_customers))
]

# test_potential_transactions = identify_dataset[
#     (identify_dataset['month'].isin(analysis_test_month)) &
#     (identify_dataset['stock_code'].isin(potential_stock_codes)) &
#     (identify_dataset['customer_id'].isin(train_target_customers))
# ]


labeled_return_customers =  labeled_data[labeled_data['invoice_type']=='purchase']['customer_id'].unique()

history_data['label'] = np.where(history_data['customer_id'].isin(labeled_return_customers),0,1)
```

```python
reference_date = history_data['invoice_date'].max() + pd.Timedelta(days=1)

rfm = history_data.groupby('customer_id').agg({
    'invoice_date': lambda x: (reference_date - x.max()).days,  # Recency
    'invoice': 'nunique',                                       # Frequency
    'revenue': 'sum'                                            # Monetary
}).reset_index()

rfm.columns = ['customer_id', 'recency', 'frequency', 'monetary']

rfm['recency_rank'] = rfm['recency'].rank(method='first')
rfm['recency_score'] = pd.qcut(rfm['recency_rank'], 5, labels=[5,4,3,2,1])

rfm['monetary_rank'] = rfm['monetary'].rank(method='first')
rfm['monetary_score'] = pd.qcut(rfm['monetary_rank'], 5, labels=[1,2,3,4,5])


rfm['frequency_rank'] = rfm['frequency'].rank(method='first')
rfm['frequency_score'] = pd.qcut(rfm['frequency_rank'], 5, labels=[1, 2, 3, 4, 5])


rfm['recency_score'] = rfm['recency_score'].astype(int)
rfm['frequency_score'] = rfm['frequency_score'].astype(int)
rfm['monetary_score'] = rfm['monetary_score'].astype(int)

rfm['fm_score'] = round((rfm['frequency_score'] + rfm['monetary_score'])/2,0)
```

```python
rfm['category'] = rfm.apply(categorize, axis=1)
```
